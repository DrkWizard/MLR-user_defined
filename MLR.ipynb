### B=(X'X)inv*(X'Y)

```python
X_t=np.transpose(X)
```

```python
X_t_X=np.matmul(X_t,X)
```

```python
X_t_X_inv=np.linalg.inv(X_t_X)
```

```python
X_t_Y=np.matmul(X_t,Y)
```

```python
B=np.matmul(X_t_X_inv,X_t_Y)
```

```python
B
```

## Model Performance

```python
y_test_=y_test.to_numpy()
```

```python
x1=list()
x2=list()
x3=list()
x4=list()
x5=list()
x6=list()
x7=list()
for i in x_test["GRE Score"]:
    x1.append(i)
for i in x_test["TOEFL Score"]:
    x2.append(i)
for i in x_test["University Rating"]:
    x3.append(i)
for i in x_test["SOP"]:
    x4.append(i)
for i in x_test["LOR "]:
    x5.append(i)
for i in x_test["CGPA"]:
    x6.append(i)
for i in x_test["Research"]:
    x7.append(i)
    
```

```python
B[1]
```

```python
y_pred=[(B[0] + B[1]*(X1) + B[2]*(X2) + B[3]*(X3) + B[4]*(X4) + B[5]*(X5) + B[6]*(X6) + B[7]*(X7) )for X1,X2,X3,X4,X5,X6,X7 in zip(x1,x2,x3,x4,x5,x6,x7)]
```

```python
residuals=[(o-p)**2 for o,p in zip(y_test_,y_pred)]
```

```python
def summation(val):
    res=0
    for i in val:
        res+=i
    return res
```

```python
summation(residuals)
```

```python
mse=summation(residuals)/x_test.shape[0]
```
